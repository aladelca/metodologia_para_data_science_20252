```python
import pandas as pd
import matplotlib.pyplot as plt

archivo_de_datos = 'Bcrp_2025.csv'


df = pd.read_csv(archivo_de_datos, header=None, skiprows=6)

# Asignar nombres a las columnas para que sean ms claras
df.columns = ['Fecha', 'Ingresos Tributarios (millones S/)']


df.set_index('Fecha', inplace=True)

# Eliminar la primera fila si es un encabezado repetido
df = df.iloc[1:]

# Crear el grfico
plt.figure(figsize=(12, 6))
df['Ingresos Tributarios (millones S/)'].plot(title='Ingresos Tributarios del Gobierno (2003-2025)')
plt.xlabel('Fecha')
plt.ylabel('Ingresos (millones de S/)')
plt.grid(True)
plt.show()

# Mostrar la lista de los ltimos 5 datos
print("\nltimos 5 datos de ingresos tributarios:")
print(df.tail())
```
